```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
import plotly.express as px 
```

```python
df = sns.load_dataset('titanic')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop(['deck'] , axis = 1 , inplace = True  )
```

```python
df
```

```python
df.dropna( axis = 0 )
```

```python
df.dropna(subset = ['embarked'] , axis = 0 , inplace = True  )
```

```python
df['age'] = df['age'].fillna(df['age'].median())
```

```python
df.info()
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.decomposition import PCA
```

```python
df.drop(['alive'] , axis = 1 , inplace = True )
```

```python
df.reset_index(inplace = True , drop = True )
```

```python
ohe = OneHotEncoder(sparse_output=False , drop = 'first')

ohe_col = ['sex' , 'embarked' , 'class' , 'who' ,'adult_male','embark_town','alone']

df_ohe = pd.DataFrame(data = ohe.fit_transform(df[ohe_col]) , columns=ohe.get_feature_names_out())

df = pd.concat([df , df_ohe] , axis = 1 ).drop(ohe_col , axis = 1)
```

```python
x , y = df.drop(['survived'] , axis = 1 ) , df['survived']
```

```python
pca = PCA(n_components=0.99)
# pca = PCA(n_components=2)
```

```python
pca.fit_transform(x)
```

```python
pca = PCA(n_components=2)
```

```python
com_x = pd.DataFrame(pca.fit_transform(x) , columns=pca.get_feature_names_out())
```
