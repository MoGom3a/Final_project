```python

import streamlit as st
import pandas as pd
import joblib

# Optional: remove unused imports
# import sklearn  # not used directly
# import category_encoders  # not used directly

st.set_page_config(layout='wide')

def get_input():
    Active_users = st.slider('Select Active Users on Site', 1, 18, 10, 1)
    Sessions = st.slider('Select Sessions', 1, 18, 10, 1)
    Event_count = st.slider('Select Event Count', 1, 22, 10, 1)
    Key_events = st.slider('Select Key Events', 1, 20, 10, 1)

    return pd.DataFrame(
        [[Active_users, Sessions, Event_count, Key_events]],
        columns=['Active users', 'Sessions', 'Event count', 'Key events']
    )

# Load input from user
test = get_input()

# Load model
try:
    svc = joblib.load('svc.h5')
    prediction = svc.predict(test)
    st.subheader("Prediction:")
    st.write(prediction[0])
except FileNotFoundError:
    st.error("Model file 'svc.h5' not found. Please make sure it's in the same directory.")
except Exception as e:
    st.error(f"An error occurred: {e}")
```
